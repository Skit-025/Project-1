## OS FOR HF_TOKE AND REQUEST FOR POST METHOD

```python
import os
import requests
```

## API END POINT WHERE REQUEST WILL BE SENT

```python
Api_url="http://router.huggingface.co/v1/chat/completions"
```

## HEADERS CARRY THE REQUEST IDENTITY AND CONTENT TYPE THAT WILL BE RETURNED

```python
headers={
    "authorization":f"Bearer {os.getenv('HF_TOKEN')}",
    'content_type':'application/json'
}
```

## LESS IMPORTANT PART CUSTOMISABLE


```python
name=input("Enter the character you wanna talk to: ")
```

## THE FUNCTION DEFINATION THAT HOLDS THE PAYLOAD

```python
def custom(message):
    #body of the engine is it's payload ESSENTIAL OR MOST IMP PART
    payload = {
    "model": "meta-llama/Meta-Llama-3-8B-Instruct",
    "messages": [
        {
            "role": "system",
        "content": f"You are {name} you are very friendly just ineract with the user properly and answer as much as you know don\'t hallucinate"
        },
        {
            "role": "user",
            "content": message
        }
    ],
    "max_tokens": 500
}
    #sents post req to huggingface api
    response=requests.post(Api_url,headers=headers,json=payload)
    print("Status:", response.status_code)# FOR DEBUGGING
    print("Raw:", response.text)# FOR DEBUGGING
    data=response.json() #IN CASE CONTENT TYPE FAILS JSONIFY AGAIN THE DATA RECEIVED
    if 'error' in data:
        return f"API Error: {data['error']}"
    
    if "choices" not in data:
        return f"API Error: {data}"

    return data["choices"][0]["message"]["content"]
    
```

```python
print(f"{name}: Hello. Type 'bye' to exit.")
while True:
    user = input("You: ")

    if user.lower() == "bye":
        break

    reply = custom(user)

    print(f"{name}: ", reply)
```
