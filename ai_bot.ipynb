```python
from IPython.core import payload
def query(message, Name="DefaultName"):
    # Decide max_token based on keywords in message
    keywords = ['generate','create','give','provide','display',"show","print"]
    max_token = 500 if any(word in message for word in keywords) else 150

    payload = {
        'model': 'metallama/Meta-llama-3-8B-Instruct',
        'messages': [
            {
                'role': 'system',
                'content': f'You are {Name}, talk to the front person just the way in which {Name} would have talked'
            },
            {
                'role': 'user',
                'content': message
            }
        ],
        'max_token': max_token
    }
    return payload
```
